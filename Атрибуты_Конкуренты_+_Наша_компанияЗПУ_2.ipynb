Каждая отдельная таблица(объект) представляет собой набор данных по остаткам и стоимости подшипниковой продукции на конкретную дату. Дата конкретной таблицы(объекта) отличаются у кажого  конкурента. Существенные различия могут быть в значениях атрибутов(количестве позиций в 'Основном обозначении') в каждой таблице(объекте) в связи с различной номенклатурной базой конкруентов. Общий период период данных 06/2023 - 06/2025, где по некоторым конкурентам могут быть данные только за один год.

**Атрибуты**

* 'Наименование' -  Полное название товара или продукции, которую предлагает компания -
* 'Бренд' - Марка или производитель товара -
* 'Основное обозначение' — Главная характеристика или тип товара, позволяющая однозначно идентифицировать его среди аналогичных позиций. +
* 'ГОСТ/ТУ' -  Указание стандарта или технических условий, по которым произведён товар (например, ГОСТ 123-45 или ТУ 6789-01) -
* 'Примечание' — Дополнительная информация или комментарии, относящиеся к товару (например, особенности, условия доставки). -
* 'Ед. измерения' — Обозначение единицы, в которой измеряется количество товара (штуки). -
* 'Количество'— Фактическое число или объём товара, который предлагается на конкретной таблицы(объекта). +
* 'Цена'— Стоимость одной единицы товара в рублях на конкретную дату таблицы(объекта). +
* 'Дата прайса' — Дата, когда был сформирован или актуализирован прайс-лист с ценой и количеством товара на конкретную дату. +
* 'Дата выгрузки' — Дата, когда данные были выгружены или сохранены из базы/системы. -
* 'Название компании' — Имя организации, предоставляющей данный товар или услугу. +
* 'Город' — Город, в котором расположена компания или склад с товаром. +
* 'Телефон'— Контактный номер компании для связи по вопросам товара или заказа. -

# **Merge Таблиц**

```python
df1 = pd.read_excel('Общая_ЗПУ.xlsx')
df2 = pd.read_excel('Общая_конкуренты.xlsx')
```

```python
df1.columns
```

```python
df2.columns
```

```python
df1['Месяц'] = pd.to_datetime(df1['Дата прайса']).dt.to_period('M')
df2['Месяц'] = pd.to_datetime(df2['Дата прайса']).dt.to_period('M')
```

```python
merged_df = pd.merge(
    df1,
    df2,
    on=['Основное обозначение', 'Месяц'],
    how='inner',
    suffixes=('_pricezpu', '_pricek')
).drop_duplicates()
```

```python
merged_df
```

```python
merged_df.to_excel('Общая.xlsx', index=False)
```

# **Себестоимость**

**Атрибуты**

* 'Основное обозначение' — Главная характеристика или тип товара, позволяющая однозначно идентифицировать его среди аналогичных позиций. +
* 'Модификация' — версия или вариант изделия, отличающийся техническими характеристиками или исполнением. +
* 'Еврообозначение' —  европейское или международное маркировочное обозначение товара, применяемое для унификации идентификации в зарубежных стандартах. -
* 'Размер' — габаритные параметры или диаметр, длина, ширина товара. +
* 'Количество роликов' — число роликов, входящих в состав изделия или комплекта. +
* 'Зазор мкм' — величина зазора между компонентами, указанная в микрометрах.+
* 'BRAND' — торговая марка или производитель товара.+
* 'Вес штука' — масса одного отдельного изделия. +
* 'Вес контракт' — общий вес продукции по контракту. +
* 'Количество в контракте' — число изделий, входящих в текущий контракт на поставку. +
* 'Цена по ФОБ в долларах' — стоимость одной единицы продукции по условиям FOB (Free On Board) в долларах США. +
* 'Себестоимость' — цена производства или закупки единицы товара. +
* 'Цена' — установленная стоимость для продажи изделия. +
* '№ контракта' — индивидуальный номер документа, по которому выполняется поставка или продажа. -
* 'Запуск контракта' — дата/информация(об оплате контракта) о начале выполнения контракта. -
* 'Поступление на склад' — дата поступления товара на склад или его приемка кладовщиком/обновление прайса в 1С. +
